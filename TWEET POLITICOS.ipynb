```python
# Importamos las librerias necesarias
import os
import pandas as pd
```

```python
# MARCELO EBRARD
tweet_count = 1000
username = "m_ebrard"
since_date = "2021-04-01"
until_date = "2022-04-01"

# Using OS library to call CLI commands in Python
os.system('snscrape --jsonl --max-results {} --since {} twitter-search  "{} until:{}"> m_ebrard.json'.format(tweet_count, since_date, username, until_date))

# Creamos dataframe con los datos
tweets_df2 = pd.read_json('m_ebrard.json', lines=True)

# Visualizamos las primeras 5 entradas del dataframe creado
tweets_df2.head()

# Exportamos el dataframe a CSV
tweets_df2.to_csv('m_ebrard.csv', sep=',', index=False)
```
